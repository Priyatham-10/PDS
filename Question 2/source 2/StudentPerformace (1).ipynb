```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the dataset
info = pd.read_csv('StudentsPerformance.csv')
```

```python
# create a bar plot for Parent level education
plt.figure(figsize=(5,5))
info['parental level of education'].value_counts().plot(kind='bar')
plt.title('parental level of education Distribution')
plt.xlabel('parental level of education')
plt.ylabel('Count')
plt.show()
```

```python
#Box plot of reading by Gender

import seaborn as dns #to enhance the visualizations and add additional functionalities
dns.boxplot(x="reading score", y="gender", data=info)
plt.xlabel("Reading_Score")
plt.ylabel("Gender")
plt.title(" Read score by gender")
plt.show()
```

```python
#Scatter plot for writing score vs reading score
import pandas as pd
import matplotlib.pyplot as plt
info = pd.read_csv('StudentsPerformance.csv')
plt.scatter(info['writing score'], info['reading score'])
plt.xlabel('writing Score')
plt.ylabel('Reading Score')
plt.title('writing Score vs Reading Score')
plt.show()
```

```python
#Histogram plot for Distribution of writing score 
import pandas as pd
import matplotlib.pyplot as plt
info = pd.read_csv('StudentsPerformance.csv')
plt.hist(info["writing score"], bins=10)
plt.xlabel("writing Score")
plt.ylabel("Frequency")
plt.title("Distribution of writing Scores")
plt.show()
```

```python
#area plot
ri = info.groupby('race/ethnicity')['reading score','writing score'].mean()
ri.plot(kind='area', stacked=False)
plt.title('Average Test Scores for Each Race/Ethnicity Category')
plt.xlabel('Race/Ethnicity Category')
plt.ylabel('Average Test Score')
plt.show()
```
